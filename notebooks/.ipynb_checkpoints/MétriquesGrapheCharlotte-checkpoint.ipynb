```python
from py2neo import Graph, authenticate
import networkx as nx
import igraph as ig
import logging
```

```python
#Authentification Neo4j
authenticate("192.168.99.100:7474", "neo4j", "pass4dbse")
graph = Graph("http://192.168.99.100:7474/db/data/")
```

```python
def getIGraph(query, graph):
    '''
    Cette fonction renvoie le graphe (au format igraph) récupéré depuis Neo4j.

    Args:

    query (string): requête Cypher à Neo4j
    graph_url (string): adresse de la base Neo4j

    Returns:

    igraph.Graph: Un objet Graph de la librairie igraph

    '''
    nodes = []
    edges = []
    for record in graph.cypher.execute(query):
        edge = {
            'source': record["nodeFrom"].properties['code'],
            'target': record["nodeTo"].properties['code']
        }
        edges.append(edge)
        nodes.append(record.nodeFrom.properties)
        nodes.append(record.nodeTo.properties)
    nodes = {v['code']: v for v in nodes}.values()  # filter duplicates
    if len(edges) == 0:
        logging.warning("no edges detected")
        raise Exception
    g = ig.Graph.DictList(nodes, edges, vertex_name_attr='code')
    return g

subgraph = getIGraph("MATCH (k:`tigr-keyword`) WHERE NOT k.blacklisted=True WITH collect(k.code) as keywordList OPTIONAL MATCH (t:`tigr-tweet`)-[:`tigr-from`]->(u:`tigr-user`) WHERE (t.language in ['fr','und', 'en', 'es'] ) AND (u.code in keywordList) AND (toInt(timestamp())-toInt(t.created_at))/(1000*60*60*24)<7 WITH keywordList,collect(distinct t.code) as tweets OPTIONAL MATCH (t:`tigr-tweet`)-[:`tigr-#about`]->(h:`tigr-hashtag`) WHERE (t.language in ['fr','und', 'en', 'es'] ) AND (('#'+ h.code) in keywordList) AND (toInt(timestamp())-toInt(t.created_at))/(1000*60*60*24)<7 WITH (tweets + collect(distinct t.code)) as tweets2 MATCH (n:`tigr-user`)<-[r1:`tigr-from`]-(t2:`tigr-tweet`)-[re1:`tigr-@mention`]->(p:`tigr-user`) WHERE (p.code <> n.code) AND (t2.code in tweets2) WITH DISTINCT n as nodeFrom,p as nodeTo, t2 order by t2.created_at RETURN nodeFrom,nodeTo", graph)
print "nombre de noeuds : "+str(len(subgraph.es))
print "nombre d'arêtes : "+str(len(subgraph.vs))
#Average clustering coefficient
acc = subgraph.transitivity_undirected()
#Diametre
diameter = subgraph.diameter()
#closeness centrality des noeuds
closeness = subgraph.closeness()
#Betweenness centrality
betweenness = subgraph.betweenness()
#Density
density = subgraph.density()
#Degré moyen
mean_degree = sum(subgraph.degree())/float(len(subgraph.degree()))
print "acc : "+str(acc)
print "diameter : "+str(diameter)
print "closeness moyenne: "+str(sum(closeness)/float(len(closeness)))
print "betweenness moyenne: "+str(sum(betweenness)/float(len(betweenness)))
print "density : "+str(density)
#Degree distribution
print subgraph.degree_distribution()
#betweenness distribution
from matplotlib import pyplot as plt
%matplotlib inline
plt.hist(subgraph.betweenness())
plt.hist(subgraph.closeness())
```
